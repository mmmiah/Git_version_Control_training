```python
# Make padas data frame after aggregration
new_atd_data = atd_date.groupby(['district', 'route'])['loc_id'].count().reset_index()
new_atd_data.rename(columns={'loc_id': 'Count'}, inplace=True)
new_atd_data
```

```python
district_counts = new_atd_data.groupby('district')['Count'].sum()
district_counts
```

```python
# using this script for tomorrow's training
```

```python
# Plot
plt.figure(figsize=(8, 5))
# Group by district and sum the counts
#district_counts = new_atd_data.groupby('district')['Count'].sum()
district_counts.plot(kind='bar', color='skyblue', edgecolor='black')

# Labels and title
plt.xlabel('District')
plt.ylabel('Total Count')
plt.title('Count by District')
plt.xticks(rotation=0)  

# Show plot
plt.show()
```

```python
district_countss = atd_date.groupby(['district'])['loc_id'].count().reset_index()
district_countss.rename(columns={'loc_id': 'Count'}, inplace=True)
district_countss
```

```python
district_countss.describe()
```

```python
# Plot with seaborn
plt.figure(figsize=(8, 5))
sns.barplot(data=district_countss, x='district', y='Count', palette='Greens')

# Labels and title
plt.xlabel('District')
plt.ylabel('Total Count')
plt.title('Count by District')
plt.xticks(rotation=0)  # Keep labels horizontal

# Show plot
plt.show()
```

```python
# Plot
plt.figure(figsize=(12, 5))
# Group by district and sum the counts
rt_counts = new_atd_data.groupby('route')['Count'].sum()
rt_counts.plot(kind='bar', color='skyblue', edgecolor='black')

# Labels and title
plt.xlabel('Route')
plt.ylabel('Total Count')
plt.title('Count by District')
plt.xticks(rotation=0)  

# Show plot
plt.show()
```

# Lets see how does geopandas work

```python
geometry = [Point(xy) for xy in zip(atd_date['longitude'], atd_date['latitude'])]
at_gdf = gpd.GeoDataFrame(atd_date, geometry=geometry)

# Set the CRS to EPSG:4326 (WGS 84)
at_gdf.set_crs(epsg=4326, inplace=True)
at_gdf_proj=at_gdf.to_crs("EPSG:3310")
at_gdf_proj = at_gdf_proj[(at_gdf_proj['district'] == 4) | (at_gdf_proj['district'] == 3)]
at_gdf_proj.head()
```

```python
# read state highway system
shs=gpd.read_file('O:\linux_mount/Jupyter/Dev_Mintu/PeMS_Coverage/SHN_Lines.geojson')
shs_proj=shs.to_crs("EPSG:3310")
shs_proj.head()
```

```python
shs_proj.crs
```

```python
fig, ax = plt.subplots(figsize = (30,25)) 
shs_proj.plot(ax=ax,color="red",edgecolor='k',linewidth = 1)
at_gdf_proj.plot(color="green",ax=ax) #Use your second dataframe
```

```python
# Now save the file as geojson and read in GIS
at_gdf_proj1 = gpd.GeoDataFrame(at_gdf_proj, geometry='geometry', crs="EPSG:3310")
# Transform CRS to EPSG:4326
summary_gdf_4326 = at_gdf_proj1.to_crs("EPSG:4326")
# Save as GeoJSON
crs = {'init': 'epsg:4326'}
final_data = gpd.GeoDataFrame(summary_gdf_4326, crs=crs, geometry = 'geometry')
final_data.to_file("AT_deployment_coverage_epsg_4326.json", driver="GeoJSON")
```
